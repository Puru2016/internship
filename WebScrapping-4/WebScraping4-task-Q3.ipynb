```python
import selenium
import pandas as pd
from selenium import webdriver
import warnings
warnings.filterwarnings('ignore')
from selenium.webdriver.common.by import By
import time

from bs4 import BeautifulSoup
import requests
import re
```

```python
driver = webdriver.Chrome()
driver.get("http://statisticstimes.com/")
```

```python
time.sleep(5)

closepopup = driver.find_element(By.XPATH,'//*[@id="top"]/div[2]/div[2]/button')
closepopup.click()

time.sleep(3)
clickoncategory = driver.find_element(By.XPATH, '/html/body/div[2]/div[1]/div[2]/div[2]/div/a[3]')
clickoncategory.click()


time.sleep(3)
clickoncategory = driver.find_element(By.XPATH, '/html/body/div[2]/div[2]/div[2]/ul/li[1]/a')
clickoncategory.click()
```

```python
url = driver.current_url
#url = 'https://www.statisticstimes.com/economy/india/indian-states-gdp.php'
page = requests.get(url)
soup = BeautifulSoup(page.text, 'html')

#print(soup)
```

```python
#table = soup.find_all('table')
#table = soup.find("table",{"class":"display dataTable"})
table = soup.find('table', id="table_id")
```

```python
contents = []

# Getting all rows
for row in table.find_all('tr')[2:]:
    contents.append(re.split('\n', row.text))
    #contents.append(row.text)
    
#contents
```
