# 1. Scrape the details of most viewed videos

```python
table = soup.find_all('table')[1]
```

```python
#print(table)
```

```python
world_titles = table.find_all('th')

#world_titles
```

```python
world_table_titles = [title.text.strip() for title in world_titles[0:6]]

print(world_table_titles)
```

```python
df =  pd.DataFrame(columns = world_table_titles)
df
```

```python
column_data = table.find_all('tr')
```

```python
for row in column_data[1:-1]:
    row_data = row.find_all('td')
    induvidual_row_data = [data.text.strip() for data in row_data]

    #print(induvidual_row_data)
    length = len(df)
    df.loc[length] = induvidual_row_data

df.sort_values(by=['Views (billions)'])
```

# Q8: Write a python program to scrape data for Top 1000 Quotes of All Time.

```python
driver = webdriver.Chrome()
driver.get("https://www.azquotes.com/")
```

```python
time.sleep(5)

clicktopquotes = driver.find_element(By.XPATH,"/html/body/div[1]/div[1]/div[1]/div/div[3]/ul/li[5]/a")
clicktopquotes.click()
```

```python
quotearray = []
authorarray = []
quotetypearry = []
```

```python
start=0
end=9

for pages in range(start,end):
    quotes = driver.find_elements(By.XPATH,'//a[@class="title"]')

    for i in quotes:
        quotevalue = i.text
        quotearray.append(quotevalue)
        
    authors = driver.find_elements(By.XPATH,'//div[@class="author"]')

    for i in authors:
        authorvalue = i.text
        authorarray.append(authorvalue)
        
    types = driver.find_elements(By.XPATH,'//div[@class="tags"]')

    for i in types:
        typevalue = i.text
        quotetypearry.append(typevalue)
    
    next_button = driver.find_element(By.XPATH, '//li[@class="next"]')
    next_button.click()
    time.sleep(15)
    
    
print(len(quotearray),len(authorarray),len(quotetypearry))
```

```python
# Making Dataframe
df = pd.DataFrame({'Quotes':quotearray, 'Authors':authorarray, 'Type Of Quotes':quotetypearry})
df = df.head(1000)
df
```
