# Write a python program to scrape cricket rankings from icc-cricket.com. You have to scrape and make data frame

# A) Top 10 ODI teams in menâ€™s cricket along with the records for matches, points and rating.

```python
#!pip install bs4
#!pip install requests

from bs4 import BeautifulSoup
import requests
import pandas as pd
import re
```

```python
page = requests.get('https://www.icc-cricket.com/rankings/mens/team-rankings/odi')
soup = BeautifulSoup(page.content)
#print(soup.prettify())

#findtable = soup.find('table')
findtable = soup.find('table', {'class': 'table'})

# Getting all rows
data = []
for row in findtable.find_all('tr'):
    row_data = []
    for cell in row.find_all('td'):
        row_data.append(cell.text.replace("\n", ""))
    data.append(row_data)

#data
```

```python
# Making Dataframe
df = pd.DataFrame(data)
df.drop(index=0)
df.columns = ['S. No', 'CountryName', 'Matches', 'Points', 'Rating']
df = df.head(11)
df
```

# B) Top 10 ODI Batsmen along with the records of their team andrating.

```python
page2 = requests.get('https://www.icc-cricket.com/rankings/mens/player-rankings/odi/batting')
soup2 = BeautifulSoup(page2.content)
#print(soup2.prettify())

#findtable = soup2.find('table')
batsmanTable = soup2.find('table', {'class': 'table rankings-table'})


# Getting all rows
batsmandata = []
for i in batsmanTable.find_all('tr'):
    bman_data = []
    for cell in i.find_all('td'):
        bman_data.append(cell.text.replace("\n", ""))
    batsmandata.append(bman_data)

#batsmandata
```

```python
# Making Dataframe
my_df = pd.DataFrame(batsmandata)

my_df.drop(columns=[my_df.columns[4]], inplace=True)
my_df.drop(index=my_df.index[0], axis=0, inplace=True)

my_df.columns = ['S. No', 'Player Name', 'TEam', 'Rating']
my_df = my_df.head(10)
my_df
```
